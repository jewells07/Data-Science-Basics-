```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Alfred'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                      "born": [pd.NaT, pd.Timestamp('1940-04-25'),
                              pd.NaT]})
```

```python
df.head()
```

```python
df.dropna()
```

If all values in column is "NaN" then and only drop

```python
df.dropna(how='all', axis = 1)
```

```python
df.drop_duplicates(subset=['name'], keep="last")
```

```python
df.drop_duplicates(subset=['name'], keep="first") # By Default its "first"
```

```python
df.drop_duplicates(subset=['name'], keep=False)
```

```python
df.shape
```

```python
df.info()
```

```python
df['toy'].value_counts(dropna=False)
```

```python
df['toy'].value_counts(dropna=True)
```

```python
df.notnull()
```

```python
df.isnull()
```

```python
# create a dataframe which contains only integers with 3 rows and 2 columns
# Run following dataframe method on them:
df.describe()
df.mean()
df.corr()
df.count()
df.max()
df.min()
df.median()
df.std()
```
